```python
df['Age_of_property'] = 2025 - df['Year_Built']
df.drop('Year_Built', axis = 1, inplace = True)
df.head()
```

```python
df['amenity_count'] = df['Amenities'].apply(lambda x :len(str(x).split(',')))
df['amenity_count'].describe()
```

```python
#calculating score for deciding whether it is a good investment
good_invest_df = pd.DataFrame()
```

```python
cols_leq_median = ['Price_in_Lakhs','Age_of_Property','price_per_sqft']

for col in cols_leq_median:
    good_invest_df[f'good_{col}'] = (df[col] <= df[col].median()).astype(int)

good_invest_df['good_amenities'] = (df['amenity_count'] >= df['amenity_count'].median()).astype(int)
good_invest_df['good_local_price'] = (df['Price_in_Lakhs'] <= df['local_avg_price']).astype(int)
good_invest_df['good_bhk'] = (df['BHK'] >= 3).astype(int)
```

```python
good_invest_df.columns
```

```python
good_invest_df['score'] = good_invest_df.sum(axis=1)
```

```python
good_invest_df.describe()
```

```python
df['good_investment'] = (good_invest_df['score']>3).astype(int)
```

```python
# calculating rate to predict future price from domain knowledge
def calculate_score(row):
    score = 0
    #price_per_sqft
    if row['price_per_sqft'] < 3000:
        score += 0
    elif 3000 <= row['price_per_sqft'] < 12000:
        score += 1
    else:
        score += 2

    #Age_of_Property 
    age = row['Age_of_Property']
    if age > 20:
        score += 0
    elif 10 <= age <= 20:
        score += 1
    else:
        score += 2

    #Nearby_Schools
    schools = row['Nearby_Schools']
    if schools == 0:
        score += 0
    elif 1 <= schools <= 3:
        score += 1
    else:
        score += 2

    #Nearby_Hospitals
    hospitals = row['Nearby_Hospitals']
    if hospitals == 0:
        score += 0
    elif 1 <= hospitals <= 3:
        score += 1
    else:
        score += 2

    #Size_in_SqFt
    sqft = row['Size_in_SqFt']
    if sqft < 800:
        score += 0
    elif 800 <= sqft <= 2500:
        score += 1
    else:
        score += 2

    #BHK
    bhk = row['BHK']
    if bhk <= 1:
        score += 0
    elif bhk == 2:
        score += 1
    else:
        score += 2

    #Amenity Count
    amenities = row['amenity_count']
    if amenities < 3:
        score += 0
    elif 3 <= amenities <= 5:
        score += 1
    else:
        score += 2

    #good_investment flag
    if row['good_investment'] == 1:
        score += 2

    return score


# -------------------------
# Apply score & compute rate
# -------------------------

df['score'] = df.apply(calculate_score, axis=1)

# Convert score  annual rate (4% to 12%)
df['annual_rate'] = 0.04 + (df['score'] / 16) * 0.08

# Future price after 5 years
df['future_price_5yr'] = df['Price_in_Lakhs'] * (1 + df['annual_rate'])**5
```

```python
import pandas as pd
import numpy as np

# Models
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from xgboost import XGBClassifier, XGBRegressor

# ML metrics
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, roc_auc_score,
    mean_squared_error, mean_absolute_error, r2_score
)

from sklearn.model_selection import train_test_split

# MLflow
import mlflow
import mlflow.sklearn
import mlflow.xgboost
import xgboost as xgb
```

```python
df = pd.read_csv("cleaned_housing_data.csv")
df.head()
```

```python
df = pd.read_csv("cleaned_housing_data.csv")

# Classification target
y_clf = df["good_investment"]
X_clf = df.drop(columns=["good_investment", "future_price_5yr"])

# Regression targetb
y_reg = df["future_price_5yr"]
X_reg = df.drop(columns=["good_investment", "future_price_5yr"])

# Train-test split
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
def evaluate_classification(model, X_test, y_test):
    preds = model.predict(X_test)
    prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    metrics = {
        "accuracy": accuracy_score(y_test, preds),
        "precision": precision_score(y_test, preds),
        "recall": recall_score(y_test, preds),
        "roc_auc": roc_auc_score(y_test, prob) if prob is not None else None
    }
    return metrics
```

```python
def evaluate_regression(model, X_test, y_test):
    preds = model.predict(X_test)
    mse = mean_squared_error(y_test,preds)
    rmse = np.sqrt(mse)
    metrics = {
        "rmse": rmse,
        "mae": mean_absolute_error(y_test, preds),
        "r2": r2_score(y_test, preds)
    }
    return metrics
```

```python
mlflow.set_tracking_uri("file:///C:/Users/Harshith Raj/mlruns")
```

```python
mlflow.set_experiment("Good_Investment_Classification")

classifiers = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForestClassifier": RandomForestClassifier(n_estimators=200, max_depth=8),
    "XGBClassifier": XGBClassifier(
        n_estimators=300, max_depth=5, learning_rate=0.1, subsample=0.8
    )
}

for model_name, model in classifiers.items():
    with mlflow.start_run(run_name=model_name):
        
        # Train
        model.fit(Xc_train, yc_train)
        
        # Evaluate
        metrics = evaluate_classification(model, Xc_test, yc_test)
        
        # Log params (only hyperparameters)
        if hasattr(model, "get_params"):
            mlflow.log_params(model.get_params())
        
        # Log metrics
        mlflow.log_metrics(metrics)
        
        # Log model
        if "XGB" in model_name:
            mlflow.xgboost.log_model(model, "model")
        else:
            mlflow.sklearn.log_model(model, "model")
        
        print(f"{model_name} Metrics:", metrics)
```

```python
mlflow.set_experiment("Future_Price_Regression")

regressors = {
    "LinearRegression": LinearRegression(),
    "RandomForestRegressor": RandomForestRegressor(n_estimators=200, max_depth=8),
    "XGBRegressor": XGBRegressor(
        n_estimators=300, max_depth=5, learning_rate=0.1, subsample=0.8
    )
}

for model_name, model in regressors.items():
    with mlflow.start_run(run_name=model_name):
        
        # Train
        model.fit(Xr_train, yr_train)
        
        # Evaluate
        metrics = evaluate_regression(model, Xr_test, yr_test)
        
        # Log params
        if hasattr(model, "get_params"):
            mlflow.log_params(model.get_params())
        
        # Log metrics
        mlflow.log_metrics(metrics)
        
        # Log model
        if "XGB" in model_name:
            mlflow.xgboost.log_model(model, "model")
        else:
            mlflow.sklearn.log_model(model, "model")
        
        print(f"{model_name} Metrics:", metrics)
```
